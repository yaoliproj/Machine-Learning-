## Tensor flow+keras (iris)

```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
import tensorflow as tf
import keras
from sklearn import datasets
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.wrappers.scikit_learn import KerasClassifier, KerasRegressor
from sklearn.grid_search import GridSearchCV
```

```python
seed = 7
np.random.seed(seed)
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, stratify=iris.target)
```

```python
num_classes = 3
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
X_train.shape
```

```python
##### model selection #####
```

```python
def make_model(optimizer="adam", hidden_size1 = 4, hidden_size2 = 4):
    model = Sequential([
        Dense(hidden_size1, input_shape=(4,), activation='relu'),
        Dense(hidden_size2, activation='relu'),
        Dense(3, activation='softmax'),       
    ])
    model.compile(optimizer=optimizer, loss="categorical_crossentropy", metrics=['accuracy'])
    return model

clf = KerasClassifier(make_model)

param_grid = {'hidden_size1': [4, 8, 32, 128, 256],
              'hidden_size2': [4, 8, 32, 128, 256]
             }

grid = GridSearchCV(clf, param_grid=param_grid, cv=5)
grid.fit(X_train, y_train)
```

```python
grid.best_params_,grid.best_score_
```

#### run model with best parameters 

```python
model = Sequential([
    Dense(grid.best_params_['hidden_size1'], input_shape=(4,), activation='relu'),
    Dense(grid.best_params_['hidden_size2'], activation='relu'),
    Dense(3, activation='softmax'),
])

model.compile("adam", "categorical_crossentropy", metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=32,
                    epochs=20, verbose=1, validation_split=.1)
```

```python
model.summary()
```

```python
##### evaluation #####
score = model.evaluate(X_test, y_test, batch_size=32, verbose=0)

print("Test loss: {:.3f}".format(score[0]))
print("Test Accuracy: {:.3f}".format(score[1]))
```
